# Wahrscheinlichkeit in einem Jupyter notebook mit python und pyplot

Das war der Prompt:

> Schreibe ein Programm das die Wahrscheinlichkeit für die Ergebnisse mit 3 Würfeln visualiziert in Python.

```python
import matplotlib.pyplot as plt
from collections import Counter

def drei_wuerfel_wahrscheinlichkeiten():
    summen = []
    for w1 in range(1, 7):
        for w2 in range(1, 7):
            for w3 in range(1, 7):
                summen.append(w1 + w2 + w3)

    counts = Counter(summen)
    gesamt = 6 ** 3  # 216 mögliche Ergebnisse

    probs = {s: counts[s] / gesamt for s in range(3, 19)}
    return probs

def plot_wahrscheinlichkeiten(probs):
    summen = list(probs.keys())
    werte = list(probs.values())

    plt.bar(summen, werte, color="skyblue", edgecolor="black")
    plt.xlabel("Summe der drei Würfel")
    plt.ylabel("Wahrscheinlichkeit")
    plt.title("Wahrscheinlichkeitsverteilung: 3 Würfel")
    plt.xticks(range(3, 19))

    # Werte über die Balken schreiben
    for s, p in probs.items():
        plt.text(s, p + 0.001, f"{p:.2%}", ha="center", fontsize=8)

    plt.show()

if __name__ == "__main__":
    probs = drei_wuerfel_wahrscheinlichkeiten()
    for s, p in probs.items():
        print(f"Summe {s:2d}: {p:.2%}")
    plot_wahrscheinlichkeiten(probs)
```

```python
import math

def combinations(n, k):
  """Calculates the number of combinations (n choose k)."""
  return math.comb(n, k)

total_combinations = combinations(49, 5)

probability = 1 / total_combinations

print(f"Die Gesamtzahl der möglichen Kombinationen (4 aus 49) beträgt: {total_combinations}")
print(f"Die Wahrscheinlichkeit, 6 Richtige zu haben, beträgt: 1 zu {total_combinations} oder ungefähr {probability:.10f}")
```

```python
# Wahrscheinlichkeit für 5 Richtige + Zusatzzahl

# Anzahl der Möglichkeiten, 5 aus den 6 gezogenen Zahlen zu treffen
combinations_5_of_6 = combinations(6, 5)

# Anzahl der Möglichkeiten, die Zusatzzahl aus den verbleibenden 43 Zahlen zu treffen (die nicht gezogen wurden)
# Beachten Sie: In der deutschen Lotterie 6 aus 49 wird die Superzahl aus den Zahlen 0-9 gezogen.
# Die Berechnung hier ist für den Fall, dass die Superzahl aus den nicht gezogenen 43 Zahlen wäre.
# Für die tatsächliche deutsche Lotterie mit Superzahl 0-9 ist die Berechnung anders.

# Lassen Sie uns die Berechnung für die tatsächliche deutsche Lotterie machen:
# 5 Richtige aus 6 gezogenen Zahlen
combinations_5_of_6 = combinations(6, 5)

# 1 Richtige Superzahl aus 10 möglichen Superzahlen (0-9)
combinations_superzahl = combinations(10, 1)

# Anzahl der Möglichkeiten, 5 Richtige UND die Superzahl zu haben
# Wir müssen auch die Anzahl der Möglichkeiten berücksichtigen, 1 Zahl falsch aus den 43 nicht gezogenen zu haben.
# Die Gesamtzahl der 6 gezogenen Zahlen ist 6. Wir haben 5 richtig, also 1 falsch aus den 43 nicht gezogenen.
combinations_1_wrong_from_43 = combinations(43, 1)

# Die Anzahl der günstigen Ereignisse (5 Richtige + Superzahl) ist:
# (Anzahl der Möglichkeiten, 5 Richtige aus 6 zu haben) * (Anzahl der Möglichkeiten, 1 falsche Zahl aus 43 nicht gezogenen zu haben) * (Anzahl der Möglichkeiten, die richtige Superzahl zu haben)
# Diese Logik ist für "5 Richtige ohne Superzahl". Für "5 Richtige mit Superzahl" ist es einfacher:
# 5 Richtige aus den 6 gezogenen Zahlen
combinations_5_of_6 = combinations(6, 5)
# Die richtige Superzahl aus den 10 möglichen Superzahlen
combinations_superzahl = combinations(10, 1)

# Die Anzahl der Kombinationen für 5 Richtige mit Superzahl ist:
# (Anzahl der Möglichkeiten, 5 aus den 6 gezogenen Zahlen zu treffen) * (Anzahl der Möglichkeiten, die richtige Superzahl zu treffen)
# ABER: Wir müssen auch bedenken, dass die 6. gezogene Zahl (die wir nicht haben) eine der 43 nicht ausgewählten Zahlen sein muss.
# Es ist einfacher, es so zu sehen:
# Wir wählen 5 Zahlen aus den 6 gezogenen (combinations(6, 5))
# Wir wählen 0 Zahlen aus den 43 nicht gezogenen (combinations(43, 0)) -> Das ist immer 1
# Wir wählen die richtige Superzahl aus den 10 möglichen (combinations(10, 1))

# Anzahl der günstigen Ereignisse für 5 Richtige + Superzahl:
favorable_outcomes_5_plus_superzahl = combinations(6, 5) * combinations(10, 1)

# Gesamtzahl der Kombinationen bleibt gleich
total_combinations = combinations(49, 6) * combinations(10, 1) # Gesamtzahl der möglichen Kombinationen mit Superzahl

probability_5_plus_superzahl = favorable_outcomes_5_plus_superzahl / total_combinations

print(f"Die Anzahl der günstigen Kombinationen für 5 Richtige + Superzahl beträgt: {favorable_outcomes_5_plus_superzahl}")
print(f"Die Gesamtzahl der möglichen Kombinationen mit Superzahl beträgt: {total_combinations}")
print(f"Die Wahrscheinlichkeit für 5 Richtige + Superzahl beträgt: 1 zu {total_combinations / favorable_outcomes_5_plus_superzahl:.0f} oder ungefähr {probability_5_plus_superzahl:.10f}")
```
